Exemplo de uso:

`plotafunc(lambda x: np.sqrt(x) + np.exp(np.sin(x)), np.linspace(0, 2*np.pi, 100))`


```python
import numpy as np
import matplotlib.pyplot as plt
def plotafunc(func,x):
    y = func(x)
    
    plt.plot(x,y,'o')
```
